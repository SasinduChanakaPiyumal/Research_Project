```python
from scipy.stats import zscore

# Keep only numeric columns
numeric_df = transposed_df.select_dtypes(include=['number'])

# Now apply z-score safely
z_scores = numeric_df.apply(zscore)

# Identify anomalies
anomalies = (z_scores.abs() > 3)
```

```python
corr_matrix = transposed_df.corr(numeric_only=True)
```

```python
mask = np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
upper_tri = corr_matrix.where(mask)

# Convert to long format, drop NaNs
corr_pairs = upper_tri.stack().reset_index()
corr_pairs.columns = ['Variable 1', 'Variable 2', 'Correlation']

# Sort by absolute correlation
top_corr_pairs = corr_pairs.reindex(corr_pairs['Correlation'].abs().sort_values(ascending=False).index)

# Show top 20 pairs
print(top_corr_pairs.head(50))
```
